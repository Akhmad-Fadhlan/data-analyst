```python
import pandas as pd
import numpy as np
import seaborn as sns

mv=['na']
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/student_performance.csv",na_values=mv)
df
```

```python
df.isnull().sum()
for x in df.columns:
  print(df[x].value_counts())
```

```python
df.isnull().sum()
```

```python
data=df.interpolate()
data
```

```python
data=data.drop(["StudentID","Name"],axis=1)
data
```

```python
gdr={"Female":0,"Male":1}
ps={"High":3,"Medium":2,"Low":1}
data["Gender"]=data["Gender"].replace(gdr)
data["ParentalSupport"]=data["ParentalSupport"].replace(ps)
data
```

```python
sns.heatmap(data.corr(),annot=True)
```

```python
# sns.boxplot(df["AttendanceRate"])
# sns.boxplot(df["StudyHoursPerWeek"])
# sns.boxplot(df["PreviousGrade"])
sns.boxplot(df["FinalGrade"])
```

```python
desc=data.describe()
q1=desc.loc["25%"]
q3=desc.loc["75%"]
iqr=q3-q1

lower=q1-1.5*iqr
upper=q3+1.5*iqr

for col in data.columns:
  data=data[(data[col] > lower[col]) & (data[col] < upper[col])]
data
```

```python
from sklearn.preprocessing import MinMaxScaler

scale=MinMaxScaler().fit_transform(data[data.columns])
data=pd.DataFrame(scale,columns=[col for col in data.columns])
data
```
